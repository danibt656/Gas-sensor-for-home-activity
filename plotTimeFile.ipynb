```python
import pandas as pd

FILENAME = "timeFile1.csv"
```

```python
df = pd.read_csv(FILENAME, sep=",")
print(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["accuracy"], df["f1-score"], s=df["time"]*10)
plt.show()
```

```python
print(df.shape)

dropped = []
for i, row in df.iterrows():
    if row['time'] > 0.6 or row['accuracy'] < 0.9985:
        dropped.append(i)
        
df = df.drop(dropped)

print(df.shape)
```

```python
plt.scatter(df['accuracy'], df['f1-score'], s=df['time']*10)
plt.xlabel("Accuracy (seg)")
plt.ylabel("F1-score (seg)")
plt.title("Best configurations of hyperparameters after elimination")
plt.show()
```
